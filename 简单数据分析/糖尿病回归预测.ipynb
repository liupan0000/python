```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train,target=load_diabetes(return_X_y=True)
train.shape,target.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
def score_model(model):
    errs =[]
    for i in range(10):
        X_train,X_test,y_train,y_test=train_test_split(train,target,test_size=0.2)
        model.fit(X_train,y_train)
        y_ = model.predict(X_test)
        err = mean_squared_error(y_test,y_)
        errs.append(err)
    return np.array(errs).mean()
```

```python
linear = LinearRegression()
knn = KNeighborsRegressor(n_neighbors=6)
```

```python
score_model(linear)
```

```python
score_model(knn)
```

```python
#均方误差小的可以作为选择的算法模型

diabetes = load_diabetes()
feature_names = diabetes.feature_names
diabetes_df = DataFrame(data=train,columns=feature_names)
diabetes_df
```

```python
#整合血糖含量到总表
sweet = DataFrame(data=target,columns=['sweet'])
df = pd.concat((diabetes_df,sweet),axis=1)
df.head()
```

```python
#bmi对血糖含量的影响
bmi = df["bp"].values.reshape(-1,1)
bmi
```

```python
#n_jobs 计算是执行任务的个数
linear = LinearRegression(n_jobs=None)
linear.fit(bmi,sweet)
```

```python
#封装函数，测试不同因素对血糖的影响
def relation_ship(other,target):
    linear = LinearRegression(n_jobs=None)
    other = other.values.reshape(-1,1)
    linear.fit(other,target)
    test  = np.linspace(other.min(),other.max(),10).reshape(-1,1)
    y_ =linear.predict(test)
    
    
    plt.figure(figsize=(12,8))
    ax = plt.subplot(111)
    ax.scatter(other,target.values,color="m",label="true")
    ax.plot(test,y_,color ="k",label="test")
    ax.legend()
    ax.set_yticks([target.min(),target.max()])
    ax.set_yticklabels(["min","max"])
```

```python
relation_ship(df["bmi"],df["sweet"])
```

```python
relation_ship(df["sex"],df["sweet"])
```

```python
relation_ship(df["bp"],df["sweet"])
```
